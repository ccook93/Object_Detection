```python
# Import libraries to read in data
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.DataFrame(pd.read_csv('../object-labeling/original/Transformer_Locator-export.csv'))
df.head()
        
```

```python
# Perform the conversion
xmin2 = df.xmin*(518/5184)
xmax2 = df.xmax*(518/5184)
#print(xmin2[0:4])
ymin2 = df.ymin*(388/3888)
ymax2 = df.ymax*(388/3888)

df.head()
```

```python
# Replace the original x and y coordinates
df['xmin'] = xmin2
df['xmax'] = xmax2
df['ymax'] = ymax2
df['ymin'] = ymin2
df.head()
```
